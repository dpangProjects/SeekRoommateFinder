```python
import pandas as pd
from sklearn.cluster import AgglomerativeClustering
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn import datasets
from kmodes.kprototypes import KPrototypes
from kmodes.kmodes import KModes
from sklearn import preprocessing
import numpy as np
from sklearn.cluster import KMeans
```

```python
df = pd.read_csv('userdata.csv')
```

```python
df
```

```python
i1 = []
i2 = []
i3 = []
for i, row in df.iterrows():
    interests = (row["Describe Yourself"]).split(",")
    i1.append(interests[0])
    i2.append(interests[1])
    i3.append(interests[2])
    
```

```python
df['Interest 1'] = i1
df['Interest 2'] = i2
df['Interest 3'] = i3
df.drop(columns=['Describe Yourself'])
```

```python
cost = []
K = range(1,10)
for num_clusters in list(K):
    kmode = KModes(n_clusters=num_clusters, init = "random", n_init = 20, verbose=1)
    kmode.fit_predict(df)
    cost.append(kmode.cost_)
```

```python
kmode = KModes(n_clusters=5, init = "random", n_init = 20, verbose=1)
clusters = kmode.fit_predict(df)
```

```python
df.insert(0, "Cluster", clusters, True)
df
```

```python
plt.plot(K, cost, 'bx-')
plt.xlabel('k clusters')
plt.ylabel('Cost')
plt.title('Elbow Method For Optimal k')
plt.show()
```
